```python
import requests
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
soup = BeautifulSoup(page.content, 'lxml')
```

```python
table_data = soup.find_all("div", {"class": "mw-parser-output"})
```

```python
df = pd.DataFrame(columns=['Postcode', 'Borough', 'Neighbourood'])
for mytable in table_data:
    table_body = mytable.find('tbody')
    rows = table_body.find_all('tr')
    ix=0
    for tr in rows:
        cols = tr.find_all('td')
        n=0
        Postcode=''
        Borough=''
        Neighbourood=''
        for j in cols:
            if n==0:
                Postcode=(j.text).strip()
            elif n==1:
                Borough=(j.text).strip()
            elif n==2:
                Neighbourood=(j.text).strip()
                if Neighbourood=='Not assigned':
                    Neighbourood=Borough                    
            n=n+1
        if Borough != 'Not assigned' and len(Borough)>1:
            df.loc[ix]=[Postcode,Borough,Neighbourood]
            ix = ix + 1
df = df.sort_values(['Postcode','Borough','Neighbourood'], ascending=[True,True,True])
```

```python
new_df = pd.DataFrame(columns=['Postcode', 'Borough', 'Neighbourood'])
prev_code = ''
concat_neig =''
x=0
prev_Postcode = ''
prev_Borough = ''
prev_concat_val=''
for ind,row in df.iterrows():
    if prev_code==row['Postcode']:
        prev_code = str(row['Postcode'])
        concat_neig=str(concat_neig + ','+ row['Neighbourood'])
    else:
        prev_code = row['Postcode']
        concat_neig=row['Neighbourood']
    if prev_Postcode != row['Postcode']:
        if len(prev_Postcode) >1:
            new_df.loc[x]=[prev_Postcode,prev_Borough,prev_concat_val]
            x=x+1            
    prev_Postcode=row['Postcode']
    prev_Borough=row['Borough']
    prev_concat_val=concat_neig 
    
new_df    
```
